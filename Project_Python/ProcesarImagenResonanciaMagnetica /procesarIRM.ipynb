pydicom: Es la biblioteca m√°s popular para trabajar con archivos DICOM en Python. Proporciona herramientas para leer, modificar y escribir archivos DICOM.

```python
import pydicom

# Cargar un archivo DICOM
dicom_file = pydicom.dcmread('/home/camilo/LaverGit/Data_analytic_and_science_project/ProjectData_Analytic_and_Science/Project_Python/ProcesarImagenResonanciaMagnetica /AVILAMORAROBERT/2.16.840.1.113662.4.10.1.51514854.20210325.11115724.19.dcm')

# Acceder a los datos de la imagen
image_data = dicom_file.pixel_array
```

```python
import matplotlib.pyplot as plt
```

```python
# Mostrar la imagen
plt.imshow(image_data, cmap='gray')
plt.show()
```

SimpleITK: Es una biblioteca m√°s avanzada que no solo permite leer archivos DICOM, sino tambi√©n realizar procesamiento de im√°genes m√©dicas.

```python
import SimpleITK as sitk

# Cargar un archivo DICOM
dicom_reader = sitk.ImageFileReader()
dicom_reader.SetFileName('ruta/al/archivo.dcm')
image = dicom_reader.Execute()

# Convertir la imagen a un array de numpy
image_array = sitk.GetArrayFromImage(image)
```

nibabel: Aunque principalmente se utiliza para im√°genes NIfTI, tambi√©n puede manejar archivos DICOM en algunos casos.

```python
import nibabel as nib

# Cargar un archivo DICOM
dicom_image = nib.load('ruta/al/archivo.dcm')

# Acceder a los datos de la imagen
image_data = dicom_image.get_fdata()
```

dicom2nifti: Esta biblioteca es √∫til si necesitas convertir archivos DICOM a formato NIfTI.


```python
import dicom2nifti

# Convertir un archivo DICOM a NIfTI
dicom2nifti.convert_directory('ruta/al/directorio_dicom', 'ruta/al/directorio_nifti')
```

Ejemplo b√°sico con SimpleITK:

```python
import SimpleITK as sitk
import matplotlib.pyplot as plt

# Cargar el archivo DICOM
dicom_reader = sitk.ImageFileReader()
dicom_reader.SetFileName('/home/camilo/LaverGit/Data_analytic_and_science_project/ProjectData_Analytic_and_Science/Project_Python/ProcesarImagenResonanciaMagnetica /AVILAMORAROBERT/2.16.840.1.113662.4.10.1.51514854.20210325.11115724.19.dcm')
image = dicom_reader.Execute()

# Convertir la imagen a un array de numpy
image_array = sitk.GetArrayFromImage(image)

# Mostrar la imagen
plt.imshow(image_array[0], cmap='gray')  # Si es una serie de im√°genes, selecciona la primera
plt.show()
```
